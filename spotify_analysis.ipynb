```python
import pandas as pd

df = pd.read_csv("spotify_data.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['popularity'], bins=30)

plt.title("Distribution of Song Popularity")

plt.show()
```

```python
sns.scatterplot(x='danceability', y='energy', data=df)

plt.title("Danceability vs Energy")

plt.show()
```

```python
df['track_genre'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Genres on Spotify")

plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap='coolwarm')

plt.title("Audio Feature Correlation")

plt.show()
```

## Key Insights

- Songs with higher energy levels tend to have higher loudness values.
- Certain genres dominate the Spotify music catalog.
- Danceability and energy are moderately correlated.
- Popularity varies widely across tracks.
